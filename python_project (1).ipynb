```python
import random
from tkinter import Tk, Label, Entry, Button, messagebox

class OTPVerificationSystem:
    def __init__(self):
        self.generated_otp = None

    def generate_otp(self):
        """Generate a 6-digit OTP."""
        self.generated_otp = str(random.randint(100000, 999999))

    def verify_otp(self, entered_otp):
        """Verify if the entered OTP matches the generated OTP."""
        return entered_otp == self.generated_otp

class OTPVerificationGUI:
    def __init__(self, master):
        self.master = master
        self.master.title("OTP Verification System")

        self.otp_system = OTPVerificationSystem()

        self.label_email = Label(master, text="Enter your email address:")
        self.label_email.pack()

        self.entry_email = Entry(master)
        self.entry_email.pack()
        
        self.generate_button = Button(master, text="Generate OTP", command=self.generate_otp)
        self.generate_button.pack()

        self.label_otp = Label(master, text="Enter the 6-digit OTP:")
        self.label_otp.pack()

        self.entry_otp = Entry(master)
        self.entry_otp.pack()

      

        self.verify_button = Button(master, text="Verify OTP", command=self.verify_otp)
        self.verify_button.pack()

    def generate_otp(self):
        email = self.entry_email.get()
        if email:
            self.otp_system.generate_otp()
            messagebox.showinfo("OTP Generated", f"OTP generated: {self.otp_system.generated_otp}")
        else:
            messagebox.showerror("Error", "Please enter your email address.")

    def verify_otp(self):
        entered_otp = self.entry_otp.get()
        if entered_otp:
            if self.otp_system.verify_otp(entered_otp):
                messagebox.showinfo("Success", "Access granted!")
            else:
                messagebox.showerror("Error", "Access denied. Please retry.")
        else:
            messagebox.showerror("Error", "Please enter the OTP.")

def main():
    root = Tk()
    app = OTPVerificationGUI(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
